```python
data['phone_one'] = data['phone'].str.split('-').str[0]
data['phone_two'] = data['phone'].str.split('-').str[1]
```

```python
data = data.drop(['phone'],axis = 1)
```

```python
data.head()
```

```python
from sklearn.preprocessing import LabelEncoder
model = LabelEncoder()
```

```python
data.state = model.fit_transform(data.state)
data.ip = model.fit_transform(data.ip)
data.vmailplan = model.fit_transform(data.vmailplan)
data.churn = model.fit_transform(data.churn)
```

```python
y = data.churn
x = data.iloc[:,:21]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.15,random_state = 20)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

```python
predict = model.predict(x_test)
```

```python
predict
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,predict)
```
